```python
import numpy as np
```

```python
class Perceptron:
    def __init__(self, input_size, learning_rate=0.1, epochs=100):
        self.weights = np.random.rand(input_size)
        self.bias = np.random.rand()
        self.learning_rate = learning_rate
        self.epochs = epochs

    def activation_function(self, x):
        return 1 if x >= 0 else 0

    def predict(self, x):
        return self.activation_function(np.dot(self.weights, x) + self.bias)

    def train(self, X, Y):
        for epoch in range(self.epochs):
            errors = 0
            for i in range(X.shape[0]):
                x = X[i]
                y = Y[i]
                prediction = self.predict(x)
                error = y - prediction
                errors += int(error != 0)
                self.weights += self.learning_rate * error * x
                self.bias += self.learning_rate * error
            if errors == 0:
                print("Converged after", epoch+1, "epochs.")
                break
            else:
                print("Epoch:", epoch+1, "Errors:", errors)
```

```python
def boolean_AND(x1, x2):
    return x1 & x2

def boolean_OR(x1, x2):
    return x1 | x2

def boolean_NAND(x1, x2):
    return ~(x1 & x2)

def boolean_XOR(x1, x2):
    return x1 ^ x2
```

```python
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])
```

```python
Y_AND = np.array([boolean_AND(x[0], x[1]) for x in X])

Y_OR = np.array([boolean_OR(x[0], x[1]) for x in X])

Y_NAND = np.array([boolean_NAND(x[0], x[1]) for x in X])

Y_XOR = np.array([boolean_XOR(x[0], x[1]) for x in X])
```

```python
print("Training Perceptron for AND function:")
perceptron_AND = Perceptron(input_size=2)
perceptron_AND.train(X, Y_AND)
print("Weights after training:", perceptron_AND.weights, "Bias:", perceptron_AND.bias)
print()
```

```python
print("Training Perceptron for OR function:")
perceptron_OR = Perceptron(input_size=2)
perceptron_OR.train(X, Y_OR)
print("Weights after training:", perceptron_OR.weights, "Bias:", perceptron_OR.bias)
print()
```

```python
print("Training Perceptron for NAND function:")
perceptron_NAND = Perceptron(input_size=2)
perceptron_NAND.train(X, Y_NAND)
print("Weights after training:", perceptron_NAND.weights, "Bias:", perceptron_NAND.bias)
print()
```

```python
print("Training Perceptron for XOR function:")
perceptron_XOR = Perceptron(input_size=2)
perceptron_XOR.train(X, Y_XOR)
print("Weights after training:", perceptron_XOR.weights, "Bias:", perceptron_XOR.bias)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def gradient_descent(gradient, initial_point, learning_rate=0.001, threshold=1e-5, max_iterations=1000):
    current_point = initial_point
    iteration = 0
    while iteration < max_iterations:
        gradient_at_current = gradient(current_point)
        new_point = current_point - learning_rate * gradient_at_current
        if np.all(np.abs(new_point - current_point) < threshold):
            break
        current_point = new_point
        iteration += 1
    return current_point, iteration
```

```python
def equation1(x):
    return x**2 - 2*x + 2

def gradient1(x):
    return 2*x - 2

def equation2(x, y):
    return (1 - x)**2 + 100*(y - x**2)**2

def gradient2(x, y):
    grad_x = -2*(1 - x) - 400*x*(y - x**2)
    grad_y = 200*(y - x**2)
    return np.array([grad_x, grad_y])
```

```python
initial_point1 = 0
initial_point2 = np.array([0, 0])
```

```python
minima1, iterations1 = gradient_descent(gradient1, initial_point1)
```

```python
minima2, iterations2 = gradient_descent(lambda point: gradient2(point[0], point[1]), initial_point2)
```

```python
print("For equation 1:")
print("Global minima found at:", minima1)
print("Number of iterations taken:", iterations1)
```

```python
print("\nFor equation 2:")
print("Global minima found at:", minima2)
print("Number of iterations taken:", iterations2)
```

```python
x_values = np.linspace(-1, 3, 400)
y_values = np.linspace(-1, 3, 400)
X, Y = np.meshgrid(x_values, y_values)
```

```python
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(1, 2, 1)
ax1.contour(X, equation1(X), Y, levels=50)
ax1.plot(minima1, equation1(minima1), 'ro')
ax1.set_title('f(x) = x^2 - 2x + 2')
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')

ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.plot_surface(X, Y, equation2(X, Y), cmap='coolwarm')
ax2.scatter(minima2[0], minima2[1], equation2(minima2[0], minima2[1]), color='red')
ax2.set_title('f(x, y) = (1 - x)^2 + 100(y - x^2)^2')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('f(x, y)')

plt.show()
```
